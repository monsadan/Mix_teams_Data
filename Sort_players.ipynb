```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
registros = pd.read_excel('registros.xlsx', engine='openpyxl')
```

```python
print(registros.shape)
```

```python
registros.head()
```

```python
registros = pd.get_dummies(registros, columns=['Resultado'])
```

```python
# jugadores.sample(frac=0.5)
```

```python
suma = registros.groupby(['Jugador']).agg({'Gol':'sum','Gol Favor':'sum','Gol Contra':'sum',
                                          'Resultado_Empate':'sum','Resultado_Gana':'sum','Resultado_Pierde':'sum'}).reset_index()
promedio = registros.groupby(['Jugador']).agg({'Gol':'mean','Gol Favor':'mean','Gol Contra':'mean'}).reset_index()
```

```python
jugadores = suma.merge(promedio, on = 'Jugador')
```

```python
jugadores = jugadores.rename(columns={'Gol_x':'total_goles', 'Gol Favor_x':'total_goles_favor','Gol Contra_x':'total_gol_contra',
                         'Gol_y':'total_goles', 'Gol Favor_y':'total_goles_favor','Gol Contra_y':'total_gol_contra'})
```
